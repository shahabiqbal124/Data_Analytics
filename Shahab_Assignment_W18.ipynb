```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://countrycode.org/"
url
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, "html.parser")
soup
```

```python
table = soup.find("table")
table
```

```python
column_headers = table.find_all("th")
column_titles = [th.text.strip() for th in column_headers]
column_titles
```

```python
clean_columns = []
for title in column_titles:
    if title and title not in clean_columns:
        clean_columns.append(title)
```

```python
df = pd.DataFrame(columns=clean_columns)
df
```

```python
table_rows = table.find_all("tr")
table_rows
```

```python
for row in table_rows:
    cells = row.find_all("td")
    if len(cells) == len(clean_columns):
        row_data = [cell.text.strip() for cell in cells]
        df.loc[len(df)] = row_data
```

```python
df.to_csv("country_codes.csv", index=False)
```
