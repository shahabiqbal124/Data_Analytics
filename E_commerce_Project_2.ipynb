## 2. Data Cleaning

### Drop irrelevant columns (Latitude, Longitude) 

```python
df = df.drop(columns=["Latitude", "Longtitude"])
df
```

### Rename columns City Code to City & Order Priority Code to Order Priority 

```python
df = df.rename(columns={"City Code": "City","Order Priority Code": "Order Priority"})
df
```

### Check duplicate rows 

```python
df.duplicated()
```

```python
df.duplicated().value_counts()
```

### Drop duplicate records 

```python
df = df.drop_duplicates()
df
```

### Check missing and null values  

```python
df.isnull().sum()
```

```python
city_dict = {1: "Karachi",2: "Lahore",3: "Islamabad",4: "Faisalabad",5: "Quetta",6: "Peshawar"}
city_dict
```

```python
df["City"] = df["City"].map(city_dict)
```

### Fill Missing Values in Categorical Column: "City"

```python
city_mode = df['City'].mode()[0]
df['City'] = df['City'].fillna(city_mode)
```

```python
df.isnull().sum()
```

```python
median = df['Order Priority'].median()
df['Order Priority'].fillna(median, inplace=True)
```

```python
df.isnull().sum()
```

##  Exploratory Data Analysis (Key Observations)

### A. Top 5 Best Selling Products

```python
top5 = df.groupby("product")["total"].sum().sort_values(ascending=False).head(5)
```

```python
top5 
```

```python
top5_df = pd.DataFrame({
    'product': top5.index,
    'total': top5.values
})
top5_df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="product", y="total", data=top5_df, palette="viridis")
plt.title("Top 5 Best Selling Products")
plt.xlabel("Product")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

#### Headphones lead by total sales, followed by watches and mobiles. These categories appear to be high-volume sellers indicating strong consumer demand.

### B. Payment Method Distribution

```python
payment_counts = df["payment_method"].value_counts()
```

```python
payment_counts
```

```python
plt.figure(figsize=(7,7))
plt.pie(payment_counts, labels=payment_counts.index, autopct="%1.1f%%", startangle=90, colors=['blue', 'green', 'red', 'orange','pink'])
plt.title("Payment Method Distribution")
plt.axis("equal") 
plt.show()
```

#### The pie chart shows the relative use of payment channels (e.g., Easypaisa, Credit Card, JazzCash). This reveals customer payment preferences and can guide payment promotions.

### C. Monthwise Sales

```python
month_sales = df.groupby("Month")["total"].sum().reset_index()
```

```python
month_sales
```

```python
month_order = [ 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
month_sales['Month'] = pd.Categorical(month_sales['Month'], categories=month_order, ordered=True)
month_sales = month_sales.sort_values('Month').reset_index(drop=True)
month_sales
```

```python
plt.figure(figsize=(10,5))
plt.plot(month_sales["Month"], month_sales["total"], marker="o", linewidth=2)
plt.title("Monthwise Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=35)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### The line chart reveals seasonal peaks and troughs in total sales. Understanding these trends can guide inventory planning and marketing campaigns throughout the year.

### D. Show City wise sales 

```python
city_sales = df.groupby("City")["total"].sum().sort_values()
```

```python
city_sales_df = city_sales.reset_index()
city_sales_df.columns = ['City', 'total']
```

```python
city_sales_df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="total", y="City", data=city_sales_df, palette="pastel")
plt.title("City-wise Sales")
plt.xlabel("Total Sales ($)")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

#### Lahore generates the highest revenue, while other cities contribute comparatively less. This indicates regional differences in demand and provides opportunities to tailor marketing, logistics, and sales strategies by city.

### E. Monthly Sales by Category

```python
monthly_cat = df.groupby(["Month", "category"])["total"].sum().unstack().fillna(0)
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_cat = monthly_cat.reindex(month_order)
```

```python
monthly_cat
```

```python
plt.figure(figsize=(12,6))
monthly_cat.plot(kind="bar", colormap='viridis', figsize=(12,6))
plt.title("Monthly Sales by Category (Corrected)", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontsize=12, fontweight='bold')
plt.ylabel("Total Sales (Millions_$)", fontsize=12, fontweight='bold')
plt.xticks(rotation=35)  
plt.legend(title="Category")
plt.tight_layout()
plt.show()
```

#### Sales of categories change from month to month. Some months have one category selling more than others, while some months are more balanced. This helps in planning inventory and promotions for each category

```python
for product in df['product'].unique():
    categories = df[df['product'] == product]['category'].unique()
    print(f"{product}: {len(categories)} categories - {list(categories)}")
```

#### Several products are incorrectly categorized in the dataset, which can lead to misleading insights. For example: Headphones are sometimes listed under Fashion instead of Electronics. Bags appear under Electronics rather than Fashion. Watches, Shoes, and Laptops are inconsistently assigned across Fashion, Electronics, and Accessories. To ensure accurate analysis, we are correcting the category column to reflect the true product categories.

```python
correct_categories = {
    'Watch': 'Accessories',
    'Bag': 'Fashion',
    'Shoes': 'Fashion',
    'Mobile': 'Electronics', 
    'Headphones': 'Electronics',
    'Laptop': 'Electronics'
}
```

```python
df['category'] = df['product'].map(correct_categories)
for product in df['product'].unique():
    category = df[df['product'] == product]['category'].unique()[0]
    print(f"{product}: {category}")
```

```python
plt.figure(figsize=(12,6))
monthly_cat.plot(kind="bar", colormap='jet', figsize=(12,6))
plt.title("Monthly Sales by Category (Corrected)", fontsize=16, fontweight='bold')
plt.xlabel("Month", fontweight='bold')
plt.ylabel("Total Sales (Millions_$)", fontweight='bold')
plt.xticks(rotation=35)
plt.legend(title="Category")
plt.tight_layout()
plt.show()
```
