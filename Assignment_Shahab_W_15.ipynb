```python
## Question#9 _ Check if there are outliers in Engine HP and show a boxplot.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4)); plt.boxplot(df["Engine Hp"]); plt.title("Engine HP Outliers"); plt.ylabel("Horsepower")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(df["Engine Hp"])
plt.title("Engine HP Outliers")
plt.ylabel("Horsepower")
plt.show()
```

```python
Q1 = df["Engine Hp"].quantile(0.25)
Q3 = df["Engine Hp"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Engine Hp"] < lower_bound) | (df["Engine Hp"] > upper_bound)]
print(f"Number of Engine HP outliers: {outliers.shape[0]}")
```

```python
## Question 11 _ Find upper bound, lower bound, and number of outliers in Engine HP
```

```python
Q1 = df["Engine Hp"].quantile(0.25)
Q3 = df["Engine Hp"].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df[(df["Engine Hp"] < lower_bound) | (df["Engine Hp"] > upper_bound)]
num_outliers = outliers.shape[0]
```

```python
print(lower_bound),
print(upper_bound), 
print(num_outliers)
```

```python
## Question 12 _ Fill missing values in Number of Doors column with the median value then check if values are filled
```

```python
median_doors = df["Number Of Doors"].median()
df["Number Of Doors"].fillna(median_doors, inplace=True)
```

```python
print(f"Median value used for Number Of Doors: {median_doors}")
print(f"Missing values in 'Number Of Doors' after filling: {df['Number Of Doors'].isnull().sum()}")
```
