# Question #7

```python
zero_overtime = df[df["OvertimePay"] == 0]
zero_overtime
```

# Question #9

```python
avg_totalpay = df.groupby("JobTitle")["TotalPayBenefits"].mean()
print(avg_totalpay) 
```

```python
avg_totalpay_sorted = avg_totalpay.sort_values(ascending=False)
print(avg_totalpay_sorted) 
```

```python
top5_avg_totalpay = avg_totalpay_sorted.head(5)
print(top5_avg_totalpay)
```

# Question #12

```python
employees_per_job = df["JobTitle"].value_counts()
print(employees_per_job)
```

```python
with pd.ExcelWriter("Salaries_all_answers.xlsx") as writer:
    df.to_excel(writer, sheet_name="Full_Data", index=False)                   # Full dataset
    basepay.to_excel(writer, sheet_name="High_BasePay", index=False)           # Q4
    manager_df.to_excel(writer, sheet_name="Managers", index=False)            # Q5
    employee_pay_df.to_excel(writer, sheet_name="Employee_Pay", index=False)   # Q6
    zero_overtime.to_excel(writer, sheet_name="Zero_OvertimePay", index=False)  # Q7
    total_per_job.to_excel(writer, sheet_name="TotalPay_Per_Job", index=False)            # Q8
    top5_avg_totalpay.to_frame().to_excel(writer, sheet_name="Top5_Avg_TotalPay", index=True)  # Q9
    DF_11.to_excel(writer, sheet_name="Top_Employees_Per_Year", index=False)        # Q11
    employees_per_job.to_excel(writer, sheet_name="Employees_Per_JobTitle", index=False)     # Q12
```
