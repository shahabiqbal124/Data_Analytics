```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('database.sqlite')
```

```python
pd.read_sql_query('select * from sqlite_master', conn)
```

```python
pd.read_sql_query('select * from salaries', conn)
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

# Question No.1
### 1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')

```python
q = '''select jobtitle, ROUND(AVG(TotalPayBenefits), 2) as 'Average (TotalPayBenefits)' from salaries 
    where jobtitle not in ('Not Provided', 'Not provided') 
    group by jobtitle order by 2 desc'''

sql_query(q)
```

# Question No.2
### Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000

```python
q = ''' select jobtitle, overtimepay from salaries
    where overtimepay > 50000
    and jobtitle not in ('Not Provided', 'Not provided')
       order by 2 desc
'''

sql_query(q)
```

# Question No.3
### What is the total number of unique employees? exclude ('Not Provided', 'Not provided')

```python
q = ''' select count(distinct employeename) as Unique_Employees from salaries
    where employeename not in ('Not Provided', 'Not provided')
'''

sql_query(q)
```

# Question No.4
### Show The Top 10 highest Paid Employees

```python
q = ''' select employeename as Employee, totalpay as Pay from salaries
    order by 2 desc limit 10
'''

sql_query(q)
```

# Question No.5
### Show All Employees Ordered By Their TotalPay Benefits In Descending Order

```python
q = ''' select employeename as Employee, totalpaybenefits as Pay_Benefit from salaries
    where employeename not in ('Not Provided', 'Not provided')
    order by 2 desc
'''

sql_query(q)
```
